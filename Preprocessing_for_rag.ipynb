```python
import pandas as pd
import numpy as np
from IPython.display import display
import os
```

```python
folder_path = 'DATA EXPLORER CONTEST/News/'
```

```python
data = {}
for filename in sorted(os.listdir(folder_path)):
  if os.path.isfile(os.path.join(folder_path, filename)):
    print(f"Accessing file: {filename}")
    if filename[-4:]=='xlsx':
        data[filename] = pd.read_excel(os.path.join(folder_path, filename))
    else:
        data[filename] = pd.read_csv(os.path.join(folder_path, filename))
    display(data[filename].head())
```

```python
def clean_data(data_dict):
    news = data_dict.copy()
    for key, value in news.items():
        print(f"Làm sạch file dữ liệu: {key}")
        df = value.copy()
        if key=='CafeF_News_FPT_CMG.xlsx':
            df['summary'] = df['summary'].str.replace(r'_x000D_\n', '\n', regex=True)
            df['summary'] = df['summary'].str.replace(r'_x000D_', '\n', regex=True)
            df['summary'] = df['summary'].str.strip()
            df['date'] = pd.to_datetime(df['date'])
        elif key=='news_dividend_issue (FPT_CMG)_processed.csv':
            # Chuyển đổi định dạng các cột ngày
            date_cols = ['Ex-Dividend Date', 'Record Date', 'Execution Date']
            for col in date_cols:
                df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')
            # Trích giá trị cổ tức
            df['Dividend (VND)'] = df['Event Content'].str.extract(r'(\d{1,3}(?:,\d{3})*) đồng/CP')[0]
            df['Dividend (VND)'] = df['Dividend (VND)'].str.replace(',', '', regex=False).astype(float)
            # Trích xuất tỷ lệ thưởng cổ phiếu, ví dụ "20:3" -> 6.67
            ratio_df = df['Event Content'].str.extract(r'(\d+):(\d+)')
            df['Bonus Ratio'] = (
                ratio_df.dropna()
                .apply(lambda x: int(x[0])/int(x[1]), axis=1)
                if not ratio_df.isnull().all(axis=None)
                else None
            )
            df['Event Type'] = df['Event Type'].str.lower().str.strip()
        elif key=='news_internal_transactions (FPT_CMG)_processed.csv':
            # Chuyển các cột ngày sang kiểu datetime
            date_cols = [
                'Registered From Date', 'Registered To Date',
                'Executed From Date', 'Executed To Date'
            ]
            for col in date_cols:
                df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')
            # Các cột số có thể chứa dấu phẩy => loại bỏ dấu phẩy và chuyển thành float
            number_cols = [
                'Before Transaction Volume', 'Before Transaction Percentage',
                'Registered Transaction Volume', 'Executed Transaction Volume',
                'After Transaction Volume', 'After Transaction Percentage'
            ]
            for col in number_cols:
                df[col] = (
                    df[col].astype(str)
                    .str.replace(',', '', regex=False)
                    .str.replace('−', '-', regex=False)  # xử lý dấu âm đặc biệt nếu có
                    .replace(['nan', 'NA', 'None'], np.nan)
                ).astype(float)
            # Chuẩn hóa định dạng text: xóa khoảng trắng đầu/cuối
            text_cols = ['StockID', 'Transaction Type', 'Executor Name',
                         'Executor Position', 'Related Person Name',
                         'Related Person Position', 'Relation']
            
            for col in text_cols:
                df[col] = df[col].astype(str).str.strip()
                df[col] = df[col].replace('nan', 'Không rõ')
        elif key=='news_shareholder_meeting (FPT_CMG)_processed.csv':
            # Danh sách các cột ngày cần chuyển đổi kiểu dữ liệu
            date_cols = ['Ex-Rights Date', 'Record Date', 'Execution Date']
            for col in date_cols:
                df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')
            # Chuẩn hóa các cột kiểu text: xóa khoảng trắng đầu/cuối
            text_cols = ['StockID', 'Exchange', 'Event Type']
            for col in text_cols:
                df[col] = df[col].astype(str).str.strip()
        else:
            pass
        news[key] = df
    return news
```

```python
clean_data = clean_data(data)
for key, value in clean_data.items():
    print(key)
    display(value.head())
```

```python
for key, value in clean_data.items():
    print("Tên file:")
    print(key)
    print("Tên các cột dữ liệu:")
    print(value.info())
```

```python
def extract_all_features(ticker, cleaned_data_ticker):
    df_features = None; all_dates = []
    date_cols_to_check = {'CafeF_News_FPT_CMG.xlsx': ['date'], 'news_dividend_issue (FPT_CMG)_processed.csv': ['Ex-Dividend Date', 'Record Date', 'Execution Date'], 'news_internal_transactions (FPT_CMG)_processed.csv': ['Registered From Date', 'Registered To Date', 'Executed From Date', 'Executed To Date'], 'news_shareholder_meeting (FPT_CMG)_processed.csv': ['Ex-Rights Date', 'Record Date', 'Execution Date']}
    for key, df_clean in cleaned_data_ticker.items():
        if df_clean is not None and isinstance(df_clean, pd.DataFrame) and not df_clean.empty:
            for col in date_cols_to_check.get(key, []):
                if col in df_clean.columns:
                    dates_series = pd.to_datetime(df_clean[col], errors='coerce')
                    all_dates.extend(dates_series.dropna().tolist())
    print(sorted(all_dates))
    min_date, max_date = min(all_dates), max(all_dates)
    print(f"Khoảng thời gian dữ liệu cổ phiếu {ticker}: {min_date.date()} đến {max_date.date()}")
```
