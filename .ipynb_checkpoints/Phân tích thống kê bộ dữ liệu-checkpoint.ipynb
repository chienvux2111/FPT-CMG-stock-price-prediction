```python
free_products = products[products['Giá vốn']==0]
product_cat = free_products.groupby('Nhóm sản phẩm')['Sản phẩm'].count().reset_index().sort_values(by='Sản phẩm', ascending=False)
product_cat['Percentage'] = round(((product_cat['Sản phẩm']/product_cat['Sản phẩm'].sum())*100),2)
product_cat
```

```python
max_cogs = max(products['Giá vốn'])
min_cogs = min(products[products['Giá vốn']!=0]['Giá vốn'])
print("Giá vốn lớn nhất:")
display(products[products['Giá vốn']==max_cogs])
print("Giá vốn nhỏ nhất:")
display(products[products['Giá vốn']==min_cogs])
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=products[products['Giá vốn']!=0], y='Giá vốn')
plt.title('Boxplot của Giá vốn')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
not_free_products = products[products['Giá vốn']!=0]
Q1 = np.percentile(not_free_products['Giá vốn'], 25)
Q3 = np.percentile(not_free_products['Giá vốn'], 75)
# Tính IQR
IQR = Q3 - Q1
# Tính Lower Bound và Upper Bound
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
# Phát hiện giá trị ngoại lai
outliers = not_free_products[(not_free_products['Giá vốn']<lower_bound) | (not_free_products['Giá vốn']>upper_bound)]
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print("Giá trị ngoại lai:")
display(outliers)
```

```python
outlier_pro = outliers.groupby('Nhóm sản phẩm').agg({
    'Sản phẩm': 'count',            # Đếm số lượng sản phẩm
    'Giá vốn': 'mean'               # Tính giá trị trung bình của giá vốn
}).reset_index().sort_values(by='Sản phẩm', ascending=False).rename(columns={'Sản phẩm':'Số sản phẩm','Giá vốn':'Giá vốn trung bình'})
display(outlier_pro)
```

## Dữ liệu bán hàng

```python
sales = data['Dữ liệu bán hàng']
sales.info()
```

```python
sales.columns
```

```python
num_col = [col for col in list(sales.columns) if sales[col].dtype=='int64']
for col in num_col:
    print(col)
    display(sales[sales[col]==0])
```

```python
for col in num_col:
    max_val = max(sales[col])
    min_val = min(sales[sales[col]!=0][col])
    print(col)
    print(f"{col} lớn nhất:")
    display(sales[sales[col]==max_val])
    print(f"{col} nhỏ nhất (không phải =0):")
    display(sales[sales[col]==min_val])
```

```python
Q1 = np.percentile(sales['Đơn giá'], 25)
Q3 = np.percentile(sales['Đơn giá'], 75)
# Tính IQR
IQR = Q3 - Q1
# Tính Lower Bound và Upper Bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Phát hiện giá trị ngoại lai
price_outliers = sales[(sales['Đơn giá'] < lower_bound) | (sales['Đơn giá'] > upper_bound)]

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print("Giá trị ngoại lai:")
display(price_outliers)
```

```python
def merge(df):
    df = df.rename(columns = {'Mã Sản Phẩm':'Mã Sản phẩm', 'Chi nhánh':'Mã chi nhánh'})
    results = pd.merge(df, products, on='Mã Sản phẩm', how='left')
    results = pd.merge(results, clients, on='Mã KH', how='left')
    results = pd.merge(results, employees, on='Mã nhân viên bán', how='left')
    results = pd.merge(results, branches, on='Mã chi nhánh', how='left') 
    return results
sales_data = merge(sales)
display(sales_data.head())
```

```python
sales_data['profit_per_unit'] = sales_data['Đơn giá']/sales_data['Giá vốn']
print(sales_data['profit_per_unit'].unique())
sales_data = sales_data.drop(columns='profit_per_unit')
```
